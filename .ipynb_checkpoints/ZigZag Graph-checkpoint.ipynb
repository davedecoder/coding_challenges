```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution:
    def __init__(self):
        self.queue = []
        self.level_tree = {}
        
        
    def append_to_level_tree(self, level:str, node:TreeNode):
        if self.level_tree.get(level, None) is not None:
            self.level_tree[level].append(node)
        else:
            self.level_tree[level] = [node]
            
    def chunks(self, lst, size):
        for i in range(0, len(lst), size):
            yield lst[i:i + size]
        
    def zigzagGraph(self, root: TreeNode) -> List[List[int]]:
        res = []
        if root is None:
            return res
        i = level = 0
        self.queue.append({
            'level': level,
            'node': root
        })
        self.level_tree[str(level)] = [root]
        while True:
            if i >= len(self.queue):
                break
            node = self.queue[i]['node']
            level = self.queue[i]['level']
            n_level = level + 1
            
            if node.left is not None and node.left.val is not None:
                self.queue.append({
                    'level': n_level,
                    'node': node.left
                })
                self.append_to_level_tree(str(n_level), node.left)
            if node.right is not None and node.right.val is not None:    
                self.queue.append({
                    'level': n_level,
                    'node': node.right
                })
                self.append_to_level_tree(str(n_level), node.right)
            i += 1
        reverse = False
        for k in self.level_tree.keys():
            p_res = [node.val for node in self.level_tree[k] if node.val is not None]
            if len(p_res) > 0:
                if reverse:
                    p_res.reverse()
                reverse = not reverse
                res.append(p_res)
        return res
        
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []
        self.queue.append(root)
        res = [[root.val]]
        i = 0
        reverse = True
        while True:
            if i >= len(self.queue):
                break
            children_vals = []
            node = self.queue[i]
            if node.left is not None:
                self.queue.append(node.left)
                children_vals.append(node.left.val)
            if node.right is not None:
                self.queue.append(node.right)
                children_vals.append(node.right.val)
            if len(children_vals) > 0:
                res.append(children_vals)
            reverse = not reverse
            i += 1
        return res
        
    def zigzagLevelOrderSwitching(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []
        self.queue.append(root)
        res = [[root.val]]
        i = 0
        reverse = True
        while True:
            if i >= len(self.queue):
                break
            children_vals = []
            node = self.queue[i]
            if reverse:
                if node.right is not None:
                    self.queue.append(node.right)
                    children_vals.append(node.right.val)
                if node.left is not None:
                    self.queue.append(node.left)
                    children_vals.append(node.left.val)
            else:
                if node.left is not None:
                    self.queue.append(node.left)
                    children_vals.append(node.left.val)
                if node.right is not None:
                    self.queue.append(node.right)
                    children_vals.append(node.right.val)
            if len(children_vals) > 0:
                res.append(children_vals)
            reverse = not reverse
            i += 1
        return res
    
    def zigzagNewLevelOrder(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []
        i = 0
        reverse = True
        res = [[root.val]]
        self.queue.append({
                    "node":root,
                    "reverse": reverse
                })
        while True:
            if i >= len(self.queue):
                break
            node = self.queue[i]["node"]
            reverse = not reverse
            if node.left is not None:                
                self.queue.append({
                    "node":node.left,
                    "reverse": reverse
                })
            reverse = not reverse
            if node.right is not None:                
                self.queue.append({
                    "node":node.right,
                    "reverse": reverse
                })
            i += 1
        for n_node in self.queue:
            children_vals = []
            if n_node["reverse"] is True:
                node = n_node["node"]
                if node.right is not None:
                    if node.right.val is not None:
                        children_vals.append(node.right.val)
                if node.left is not None:
                    if node.left.val is not None:
                        children_vals.append(node.left.val)
            else:
                node = n_node["node"]
                if node.left is not None:
                    if node.left.val is not None:
                        children_vals.append(node.left.val)
                if node.right is not None:
                    if node.right.val is not None:
                        children_vals.append(node.right.val)
            if len(children_vals) > 0:
                res.append(children_vals)
        return res
    
    def zigzagNewNewLevelOrder(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []
        i = 0
        reverse = True
        res = [[root.val]]
        self.queue.append(root)
        while True:
            if i >= len(self.queue):
                break
            node = self.queue[i]
            if node.left is not None:                
                self.queue.append(node.left)
            if node.right is not None:                
                self.queue.append(node.right)
            i += 1
        for node in self.queue:
            children_vals = []
            if node.right is not None or node.left is not None:
                if reverse is True:
                    if node.right is not None:
                        if node.right.val is not None:
                            children_vals.append(node.right.val)
                    if node.left is not None:
                        if node.left.val is not None:
                            children_vals.append(node.left.val)
                else:
                    if node.left is not None:
                        if node.left.val is not None:
                            children_vals.append(node.left.val)
                    if node.right is not None:
                        if node.right.val is not None:
                            children_vals.append(node.right.val)
                
            if len(children_vals) > 0:
                reverse = not reverse
                res.append(children_vals)
        return res
        
    
    def nZigZagrder(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []
        i = 0
        reverse = True
        res = [[root.val]]
        self.queue.append(root)
        while True:
            if i >= len(self.queue):
                break
            node = self.queue[i]
            if node.left is not None:                
                self.queue.append(node.left)
            if node.right is not None:                
                self.queue.append(node.right)
            i += 1
        for node in self.queue:
            children_vals = []
            
            if len(children_vals) > 0:
                reverse = not reverse
                res.append(children_vals)
        return res
    
    
    def orderedList(self, root: TreeNode) -> List[str]:
        self.queue.append(root)
        i = 0
        while True:
            if i >= len(self.queue):
                break
            node = self.queue[i]
            if node.left is not None:
                self.queue.append(node.left)
            if node.right is not None:
                self.queue.append(node.right)
            i += 1
        res = []
        for node in self.queue:
            res.append(node.val)
        return res
```
