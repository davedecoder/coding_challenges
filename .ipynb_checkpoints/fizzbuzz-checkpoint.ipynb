```python
def fizzbuzz(limit):
    x = 0
    my_limit = abs(limit)
    while x < my_limit:
        yield x
        x += 1
```

```python
my_fizz_buzz = fizzbuzz(100)

for i in my_fizz_buzz:
    if i % (3 * 5) == 0:
        print(i, 'FizzBuzz')
    elif i % 5 == 0:
        print(i, 'Buzz')
    elif i % 3 == 0:
        print(i, 'Fizz')
    
```

```python
my_fizz_buzz = fizzbuzz(100)

for i in my_fizz_buzz:
    string = ""
    if i % 3 == 0:
        string = string + "Fizz"
    if i % 5 == 0:
        string = string + "Buzz"
    if i % 5 != 0 and i % 3 != 0:
        string = string + str(i)
    print(string)
```
